# 如何用merge/join将不同数据集合并

### 首先是merge的方法
##### concat太简单了不写了

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/Users/chenrong/Desktop/数据合并练习/语文成绩.csv")
```

```python
df2 = pd.read_csv("/Users/chenrong/Desktop/数据合并练习/数学英语成绩.csv")
```

```python
df3 = pd.read_csv("/Users/chenrong/Desktop/数据合并练习/学籍号.csv")
```

```python
df1 = df1.dropna(axis=1)
df1
```

```python
df2 = df2.dropna(axis=1)
df2
```

```python
df3 = df3.dropna(axis=1)
df3
```

这里是用学号作为索引，自动将两个表格合并到一起

```python
students_marks = pd.merge(df1,df2,on="学号")
students_marks
```

学号和学籍号是一个东西，如何不重命名列明就直接用他们来merge呢？

```python
final_result = pd.merge(students_marks,df3,left_on=["学号"],right_on=["学籍号"])
final_result
```

```python
final_result.drop("学籍号",axis = 1)
```

### 然后是join的方法

```python
consumer_info = pd.read_csv("/Users/chenrong/Desktop/客户信息.csv")
product_info=pd.read_csv("/Users/chenrong/Desktop/产品和数量.csv")
```

```python
consumer_info = consumer_info.dropna(axis=1)
product_info = product_info.dropna(axis=1)
```

```python
consumer_info
```

```python
product_info
```

```python
consumer_info.join(product_info,how = "inner",lsuffix = "客户ID-客户",rsuffix = "客户ID-产品")
#由于join是一种用位置iindex来拼接的方法，所以一旦拼接后有两列一样，就需要对重名列进行重命名
```
